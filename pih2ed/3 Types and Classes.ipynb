## 3.1 Basic concpets

A *type* is a collection of related values.

````haskell
:type False
:type True
:type not
````

````haskell
:type not False
:type not True
:type not (not False)
````

In Haskell, every expression must have a type, which is calculated priror to evaluating the expression by a process called *type inference*.

$\displaystyle
\frac{\mathtt{f ~::~ A \to B}\qquad\mathtt{e ~::~ A}}{\mathtt{f~~e ~::~ B}}
$

````haskell
:type not 3
````

## 3.3 List types

A *list* is a sequence of *elements* of the same type.

````haskell
[False,True,False] :: [Bool]
['a','b','c','d'] :: [Char]
["One","Two","Three"] :: [String]
````

````haskell
:type [False,True,False]
:type ['a','b','c','d']
:type ["One","Two","Three"]
````

````haskell
:type [False]
:type ['a']
:type [[]]
:type []
````

````haskell
:type [False,True]
:type [False,True,False]
````

````haskell
:type [['a','b'],['c','d','e']]
````

There is no restriction that a list must have a finite length.

## 3.4 Tuple types

A *tuple* is a finite sequence of *components* of possibly differrent types.

````haskell
:type (False, True)
:type (False, 'a', True)
:type ("Yes", True, 'a')
````

````haskell
:type ('a', (False, 'b'))
:type (['a','b'], [False,True])
:type [('a',False), ('b',True)]
````

## 3.6 Curried functions

````haskell
add :: (Int,Int) -> Int
add (x, y) = x + y

add' :: Int -> (Int -> Int)
add' x y = x + y
````

```haskell
Int -> (Int -> Int)
```

is same as
```haskell
Int -> Int -> Int
```

````haskell
:type add'
````

````haskell
:type (add' 2)
````

````haskell
(add' 2) 3 == add' 2 3
````

````haskell
mult :: Int -> (Int -> (Int -> Int))
mult x y z = x * y * z
````

```haskell
Int -> (Int -> (Int -> Int))
```

is same as
```haskell
Int -> Int -> Int -> Int
```

````haskell
:type mult
````

````haskell
((mult 2) 3) 4 == mult 2 3 4
````

## 3.8 Overloaded types

````haskell
1 + 2
````

````haskell
1.0 + 2.0
````

````haskell
:type (+)
````

`Num a` is a *class constraint* that constrains `a` to be any instance of the class `Num` of numeric types.

A type that contains one or more class constraints is called *overloaded*, as is an expression with such a type.

````haskell
:type (*)
:type negate
:type abs
:type 3
````

## 3.9 Basic classes

A *class* is a collection of types that support certain overloaded operations called *methods*.

### Eq
equality types

````haskell
True == False
````

````haskell
False == False
````

````haskell
'a' == 'b'
'a' == 'a'
````

````haskell
"abc" == "abc"
"abc" == "abcd"
````

````haskell
[1,2] == [1,2]
[1,2] == [1,2,3]
````

````haskell
[sin,cos] == [tan,sin]
````

````haskell
('a',False) == ('a',False)
('a',True) == ('a',False)
('a',True) == ('b',True)
````

````haskell
(sin,False) == (cos,True)
````

### Ord
ordered types

````haskell
False < True
````

````haskell
min 'a' 'b'
````

Strings, lists, and tuples are ordered *lexicographically*.

````haskell
"elegant" < "elephant"
````

````haskell
[1,2,3] < [1,2]
````

````haskell
('a',2) < ('b',1)
````

````haskell
('a',2) < ('a',1)
````
