## 소수(prime number) 리스트

`````haskell
factors n = [x | x<-[1..n], mod n x == 0]
`````

`````haskell
factors 7
factors 15
`````

`````haskell
prime n = factors n == [1,n]
`````

`````haskell
prime 7
prime 15
`````

`````haskell
primesSlow :: [Integer]
primesSlow = filter prime [1..]
`````

`````haskell
-- https://wiki.haskell.org/Timing_computations
import Text.Printf
import Control.Exception
import System.CPUTime

time :: IO t -> IO t
time a = do
    start <- getCPUTime
    v <- a
    end   <- getCPUTime
    let diff = (fromIntegral (end - start)) / (10^12)
    printf "Computation time: %0.3f sec\n" (diff :: Double)
    return v
`````

`````haskell
time . print . take 1000 $ primesSlow
`````

`````haskell
time . print . length . take 2500 $ primesSlow
`````

## 에라토스테네스의 체

`````haskell
primes :: [Integer]
primes = sieve [2..]

sieve (p:ps) = p : sieve [x | x<-ps, mod x p /= 0]
`````

## 쌍둥이 소수(twin prime)
차이가 2인 두 소수 (예: 3과 5, 5와 7)

참고로 쌍둥이 소수가 무한히 많은지는 2019년 현재까지 수학의 미해결 난제

https://en.wikipedia.org/wiki/Twin_prime

`````haskell
twin (x,y) = x+2==y
`````

`````haskell
twins = filter twin (zip primes (tail primes))
`````

`````haskell
take 100 twins
`````

`````haskell
:type zip
zip [1,2,3,4] ['a','b','c']
zip [1..10] [1..10]
zip [1..10] (tail [1..10])
`````

## 소인수분해 (2점)
소인수분해 함수 `factorize`를 앞서 정의된 `primes`를 이용해 정의하는데
아래와 같이 `facto`라는 함수를 통해 정의하려 한다.
아래 `facto` 함수의 정의에서 `undefined`로 남겨진 부분들을 완성하라.
(힌트: 표준라이브러리에서 제공하는 `div` 함수를 활용하라.)

```haskell
factorize 55800 == [2,2,2,3,3,5,5,31]
````

`````haskell
factorize :: Integer -> [Integer]
factorize n = facto n primes

facto 1 _         = []
facto n (p:ps) 
  | mod n p == 0  = undefined
  | otherwise     = undefined
`````

## 무한 다항식 (2점)
꼭 정수계수일 필요는 없지만 편의상 정수계수 다항식을 고려하자

정수계수 무한 다항식 $a_0+a_1x+a_2x^2+\ldots$을 하스켈 리스트로 다음과 같이 나타내기로 하자.

$[a_0,a_1,a_2,\ldots]$

이 때 정수계수 무한 다항식의 합을 구하는 함수 `add`와 (1점) 곱을 구하는 함수 `mul`를 (1점) 정의하라.

예컨대,
```haskell
take 7 ( add ([1,2,3]++[0,0..])
             ([1,2,3]++[0,0..]) )
           == [2,4,6,0,0,0,0]
             
take 7 ( mul ([1,2,3]++[0,0..])
             ([1,2,3]++[0,0..]) )
           == [1,4,10,12,9,0,0]
```

`````haskell
type Poly = [Integer]
`````

`````haskell
add :: Poly -> Poly -> Poly
add (a:as) (b:bs) = undefined
`````

`````haskell
mul :: Poly -> Poly -> Poly
mul (a:as) bs = undefined
`````
